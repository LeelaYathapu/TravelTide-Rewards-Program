## **Filtering the data - Defining the Cohort and Devising User Metrics**

```python
# Travel Tide currently consists of around 1.6 million users, not all have been active recently,
# we want to study the behaviour of most latest customers who were active on Traveltide hence
# we create query to filter data -defining the Cohort as per client suggestion.

query = """

-- This CTE prelimits our sessions on Elena's suggested timeframe (After Jan 4 2023)
WITH sessions AS (
  SELECT *
  FROM sessions s
  WHERE s.session_start > '2023-01-04'
),

-- This CTE returns the ids of all users with more than 7 sessions in 2023
filtered_users AS (
  SELECT user_id,
         COUNT(*) AS session_count
  FROM sessions s
  GROUP BY user_id
  HAVING COUNT(*) > 7
),

-- CTE to build the base table with user and session details
session_base AS (
  SELECT
      s.*,
      u.birthdate,
      u.gender,
      u.married,
      u.has_children,
      u.home_country,
      u.home_city,
      u.home_airport_lat,
      u.home_airport_lon,
      f.origin_airport,
      f.destination,
      f.destination_airport,
      f.seats,
      f.return_flight_booked,
      f.departure_time,
      f.return_time,
      f.checked_bags,
      f.destination_airport_lat,
      f.destination_airport_lon,
      f.trip_airline,
      f.base_fare_usd,
      h.rooms,
      CASE
  			WHEN h.nights < 0 THEN 1
  			ELSE h.nights
  		END AS nights,
      SPLIT_PART(h.hotel_name, ' ', 1) AS preferred_hotel_group,
      h.hotel_per_room_usd AS hotel_price_per_room_night_usd

  FROM sessions s
  LEFT JOIN users u ON s.user_id = u.user_id
  LEFT JOIN flights f ON s.trip_id = f.trip_id
  LEFT JOIN hotels h ON s.trip_id = h.trip_id
  WHERE s.user_id IN (SELECT user_id FROM filtered_users)
),

-- CTE to calculate distance metrics
distance_metrics AS (
  SELECT
    sb.user_id,
		ROUND(COALESCE(SUM(haversine_distance(home_airport_lat, home_airport_lon, destination_airport_lat, destination_airport_lon)), 0)::NUMERIC,2) AS total_flight_distance,
		ROUND(AVG(haversine_distance(home_airport_lat, home_airport_lon, destination_airport_lat, destination_airport_lon))::NUMERIC,2) AS avg_km_flown,
		ROUND(COALESCE(
        SUM(sb.flight_discount_amount * sb.base_fare_usd) * 1.0
        / NULLIF(SUM(haversine_distance(home_airport_lat, home_airport_lon, destination_airport_lat, destination_airport_lon)), 0),
        0
    )::NUMERIC,5) AS avg_dollars_saved_per_km
  FROM session_base sb
  GROUP BY sb.user_id
),

--  CTE to summarize user travel spend and trips
UserTravelSpendSummary AS (
  SELECT
    sb.user_id,
    ROUND(SUM(CASE WHEN sb.flight_discount= 'TRUE' AND sb.flight_discount_amount > 0 AND sb.hotel_discount = 'true' AND sb.hotel_discount_amount > 0  THEN 1 ELSE 0 END)*1.0 / NULLIF(COUNT(*), 0), 2) AS both_discount_proportion,
    ROUND(SUM(CASE WHEN sb.flight_discount='TRUE' AND sb.flight_discount_amount > 0 AND sb.hotel_discount = 'false' THEN 1 ELSE 0 END)*1.0 / NULLIF(COUNT(*), 0), 2) AS flight_discount_proportion,
    ROUND(SUM(CASE WHEN sb.hotel_discount='TRUE' AND sb.hotel_discount_amount > 0 AND sb.flight_discount ='false'  THEN 1 ELSE 0 END)*1.0 / NULLIF(COUNT(*),0), 2) AS hotel_discount_proportion,
    ROUND(SUM(sb.flight_discount_amount) * 1.0 / NULLIF(COUNT(DISTINCT sb.trip_id), 0), 2) AS flight_discount_percentage,
    ROUND(SUM(sb.hotel_discount_amount) * 1.0 / NULLIF(COUNT(DISTINCT sb.trip_id), 0), 2) AS hotel_discount_percentage,
  ROUND(AVG(
  CASE
    WHEN sb.hotel_discount = 'TRUE' THEN (sb.hotel_price_per_room_night_usd * sb.hotel_discount_amount * sb.rooms * sb.nights)
    ELSE 0
  END
),2) AS avg_dollars_saved_hotel,

COALESCE(
  SUM(
    CASE
      WHEN sb.hotel_discount = 'TRUE' THEN ROUND(sb.hotel_price_per_room_night_usd * sb.rooms * sb.nights * (1 - sb.hotel_discount_amount), 2)
      ELSE ROUND(sb.hotel_price_per_room_night_usd * sb.rooms * sb.nights, 2)
    END
  )/ NULLIF(COUNT(DISTINCT sb.trip_id), 0), 0
) AS avg_hotel_usd_spent,

COALESCE(
  SUM(
    CASE
      WHEN sb.flight_discount = 'TRUE' THEN ROUND(sb.base_fare_usd * (1 - sb.flight_discount_amount), 2)
      ELSE ROUND(sb.base_fare_usd, 2)
    END
  )/ NULLIF(COUNT(DISTINCT sb.trip_id), 0), 0
) AS avg_flight_usd_spent,

COALESCE(
  ROUND(SUM(
    CASE
      WHEN sb.hotel_discount = 'TRUE' THEN sb.hotel_price_per_room_night_usd * sb.rooms * sb.nights * (1 - sb.hotel_discount_amount)
      ELSE sb.hotel_price_per_room_night_usd * sb.rooms * sb.nights
    END
  ), 2) +
  ROUND(SUM(
    CASE
      WHEN sb.flight_discount = 'TRUE' THEN sb.base_fare_usd * (1 - sb.flight_discount_amount)
      ELSE sb.base_fare_usd
    END
  ), 2) / NULLIF(COUNT(DISTINCT sb.trip_id), 0), 0
) AS avg_total_usd_spent,

    COUNT(
      CASE
        WHEN sb.flight_booked = 'TRUE' AND sb.hotel_booked = 'FALSE' AND sb.cancellation = FALSE THEN sb.trip_id
        ELSE NULL
      END
    ) AS total_only_flights,
    COUNT(
      CASE
        WHEN sb.hotel_booked = 'TRUE' AND sb.flight_booked = 'FALSE' AND sb.cancellation = 'FALSE' THEN sb.trip_id
        ELSE NULL
      END
    ) AS total_only_hotels,
    COUNT(
      CASE
        WHEN sb.hotel_booked = 'TRUE' AND sb.flight_booked = 'TRUE' AND sb.cancellation = 'FALSE' THEN sb.trip_id
        ELSE NULL
      END
    ) AS total_together

  FROM session_base sb
  GROUP BY sb.user_id
),

-- CTE to calculate various user metrics.We want to aggregate user behaviour into metrics (a row per user)
-- This CTE contains metrics that have to do with the browsing behaviour
-- ALL SESSION within our cohort get aggregated
user_metrics AS (
  SELECT sb.user_id,
  DATE_PART('YEAR', AGE(CURRENT_DATE, sb.birthdate)) AS age,
  CASE
      WHEN EXTRACT(YEAR FROM AGE(NOW(), birthdate)) BETWEEN 15 AND 17 THEN '<18'
      WHEN EXTRACT(YEAR FROM AGE(NOW(), birthdate)) BETWEEN 18 AND 24 THEN '18-24'
      WHEN EXTRACT(YEAR FROM AGE(NOW(), birthdate)) BETWEEN 25 AND 34 THEN '25-34'
      WHEN EXTRACT(YEAR FROM AGE(NOW(), birthdate)) BETWEEN 35 AND 44 THEN '35-44'
      WHEN EXTRACT(YEAR FROM AGE(NOW(), birthdate)) BETWEEN 45 AND 54 THEN '45-54'
      WHEN EXTRACT(YEAR FROM AGE(NOW(), birthdate)) BETWEEN 55 AND 64 THEN '55-64'
      ELSE '65+'
    END AS age_group,
      COUNT(sb.session_id) AS total_sessions,
      (MAX(DATE(sb.session_end)) - MAX(DATE(sb.session_end)) FILTER(WHERE sb.trip_id IS NOT NULL)) AS recency,
      COUNT(DISTINCT sb.trip_id) AS total_trips,
      ROUND(COUNT(DISTINCT sb.trip_id) * 1.0 / NULLIF(COUNT(sb.session_id), 0), 2) AS booking_rate,
      COUNT(DISTINCT CASE WHEN sb.cancellation = 'true' THEN sb.trip_id ELSE NULL END) AS total_cancellations,

      COALESCE(
          COUNT(DISTINCT CASE WHEN sb.cancellation = 'true' THEN sb.trip_id ELSE NULL END)::FLOAT / NULLIF(COUNT(DISTINCT sb.trip_id)::FLOAT, 0),
          0
      ) AS cancellation_rate,

      AVG(sb.session_end - sb.session_start) AS avg_session_time,
      AVG(sb.return_time - sb.departure_time) AS avg_trip_duration,
      AVG(sb.page_clicks) AS average_clicks,
      SUM(sb.page_clicks::float) / COUNT(sb.session_id) AS page_clicks_rate,
      ROUND(SUM(sb.seats) * 1.0 / NULLIF(COUNT(DISTINCT sb.trip_id), 0), 2) As avg_seats,
      ROUND(SUM(sb.rooms) * 1.0 / NULLIF(COUNT(DISTINCT sb.trip_id), 0), 2) AS avg_hotel_rooms,
      ROUND(SUM(sb.checked_bags) * 1.0 / NULLIF(COUNT(DISTINCT sb.trip_id), 0), 2) AS avg_bags_checked,
      ROUND(SUM(sb.nights) * 1.0 / NULLIF(COUNT(DISTINCT sb.trip_id), 0), 2) AS avg_hotel_nights
  FROM session_base sb
  GROUP BY sb.user_id,sb.birthdate
),

--CTE to capture user demograpy data
user_demography as (SELECT user_id,married,gender,has_children,home_country FROM users)


-- Final select statement to include all aggregated metrics at the user level

SELECT
    sb.user_id,
    um.age,
    um.age_group,
    ud.gender,
    ud.married,
    ud.has_children,
    ud.home_country,
    um.total_sessions,
    um.recency,
    um.total_trips,
    uts.avg_total_usd_spent,
    uts.avg_flight_usd_spent,
    uts.avg_hotel_usd_spent,
    uts.total_together,
    uts.total_only_flights,
    uts.total_only_hotels,
   	um.booking_rate,
    um.total_cancellations,
    um.cancellation_rate,
    uts.flight_discount_percentage,
    uts.hotel_discount_percentage,
    uts.avg_dollars_saved_hotel,
    dm.avg_dollars_saved_per_km,
    uts.flight_discount_proportion,
    uts.hotel_discount_proportion,
    uts.both_discount_proportion,
    dm.avg_km_flown,
    um.page_clicks_rate,
    um.average_clicks,
    um.avg_bags_checked,
    um.avg_hotel_nights,
    um.avg_hotel_rooms,
    um.avg_seats,
    (uts.avg_dollars_saved_hotel*uts.hotel_discount_proportion*uts.hotel_discount_percentage) AS hotel_bargain_hunter_index,
    (dm.avg_dollars_saved_per_km*uts.flight_discount_proportion*uts.flight_discount_percentage) AS flight_bargain_hunter_index
FROM (
  SELECT DISTINCT user_id FROM session_base
) sb
LEFT JOIN distance_metrics dm ON sb.user_id = dm.user_id
LEFT JOIN UserTravelSpendSummary uts ON sb.user_id = uts.user_id
LEFT JOIN user_metrics um ON sb.user_id = um.user_id
LEFT JOIN user_demography ud on ud.user_id=sb.user_id;

"""
```

## **Correlation Matrix to Determine Highly Correlated Pairs**

It's important to ensure that our metrics are not highly correlated, as strong correlations can disproportionately influence the k-means clustering algorithm, potentially resulting in suboptimal outcomes. To address this, we will use a correlation matrix combined with a heatmap. This approach allows us to visually assess the strength of relationships between different metrics. By identifying and analyzing any significant correlations, we can make informed decisions about which features to include, ensuring that the clustering process focuses on relevant and independent attributes.

```python

# Calculate the correlation matrix
correlation_matrix = df_cleaned.corr()

# Plotting the heatmap
plt.figure(figsize=(14, 10))
# Create the heatmap with smaller annotation font size
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", vmin=-1, vmax=1,
            annot_kws={"size": 8})  # Adjust the "size" value to make the font smaller
plt.title('Correlation Heatmap of Scaled Metrics', fontsize=18)
plt.show()
```

```python
pairs_to_drop = set()
cols = correlation_matrix.columns
for i in range(0, correlation_matrix.shape[1]):
    for j in range(0, i+1):
        pairs_to_drop.add((cols[i], cols[j]))
```

### **Identifying Highly correlated Features**

```python
# Unstack the correlation matrix and sort by absolute value
corr_abs_unstacked = correlation_matrix.abs().unstack()
# Drop pairs to get unique (i, j) correlations (avoid redundancy)
corr_abs_unstacked = corr_abs_unstacked.drop(labels=pairs_to_drop).sort_values(ascending=False)
# Extract highly correlated pairs (absolute value greater than 0.6)
high_corr = corr_abs_unstacked[corr_abs_unstacked > 0.6]
# Display pairs of highly correlated features
print("Highly correlated pairs: ")
print(high_corr)
```

### **Removing Highly correlated pairs before fitting into Kmeans**

By eliminating highly correlated features, we can develop a model that is more efficient, interpretable, and reliable. The aim is to keep only the features that offer unique and valuable insights, ensuring variance and contributing to improved performance and easier model interpretation.


```python
#Removing highly correlated pairs
df_cleaned.drop(columns=['avg_hotel_nights','avg_dollars_saved_per_km','booking_rate','total_together','avg_dollars_saved_hotel','total_cancellations','average_clicks','avg_flight_usd_spent','page_clicks_rate','avg_seats',], inplace=True)

```

Analyzing Elbow Plot:
* From 1 to 3 clusters: There is a significant drop in inertia, meaning that each additional cluster is helping to better fit the data.
* From 3 to 5 clusters: The rate of decrease in inertia begins to slow down, indicating diminishing returns in how much better the clusters are fitting the data.
* After 5 clusters: The curve flattens out more, meaning adding more clusters provides only a small reduction in inertia.
Also our requirement from Elena states 5, so we choose 5 as optimal clusters.

## **Limitations**

K-Means Limitations

* Sensitivity to Input Data: Small changes to input data can yield wildly different results.

* Subjectivity in Cluster Interpretation: There is no clear-cut method for defining the number of clusters or for interpreting the clustering results.

* Lack of Robustness: Small changes in data parameters and effects of randomness (which cannot be completely removed) lead to variable results.

* Inability to Handle Large Datasets: Cluster analysis is computationally intensive, hence it can be prohibitive to run in large datasets.
